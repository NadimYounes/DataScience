# Plotly Bubble Map 

I use plotly here to make a bubble map graph of eartquake magnitudes in Asia. For more details on the code used, you can check the plotly docs ! 

data url:https://earthquake.usgs.gov/earthquakes/feed/v1.0/csv.php

```python
import pandas as pd #Import pandas to load csv
import plotly.plotly as py # Import plotly
import plotly.graph_objs as go # Import plotly's graph objects
from plotly import offline as pyo # Import plotly in order to load plots offline 
pyo.init_notebook_mode(connected=True) 
```

```python
df = pd.read_csv('/Users/nadimyounes/Downloads/quakes.csv') # Load CSV using pandas, I am using world earthquake data

data_quakes = [dict(type = 'scattergeo',  #I use the scattergeo plot as the eartquake data uses longitude and latitude coordinates 
lon = df['longitude'], #Specify longitude column
lat = df['latitude'], # Specify latitude column
marker = dict(size = 2**df['mag']/10,color='red'))] #Specify the marker and the size scale for the bubbles  

layout = dict(
        title = 'Asia Earthquake Magnitude Bubble Map', # Add a title to your plot 
        showlegend = False,geo = dict(scope='asia')) #Add a scope (to specify a region)
    
                    
fig = dict(data=data_quakes,layout=layout)                      
pyo.iplot(fig, validate=False, filename='d3-bubble-map-populations')
```

```python
import pandas as pd
```

```python
#Load data with movie titles and their genres from the IMDB open dataset
genres_titles = pd.read_csv('title.basics.tsv', sep='\t')
genres_titles.head(5)
```

```python
#Inspect the Genres Column 
genres_single = genres_titles[['genres']].loc[0:15]
genres_single
```

```python
#To facilitate analysis on individual genres, I wanted to split this column to have one genre per title
ind_genres = genres_single['genres'].str.split(',').apply(pd.Series)
ind_genres
```

```python
#To get all of the genres into 
genres_fin = ind_genres.stack(-1)
genres_fin.index = genres_fin.index.droplevel(-1)
genres_final = genres_fin.to_frame()
```
